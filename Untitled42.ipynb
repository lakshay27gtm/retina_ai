### Step 1: Organize the dataset

Your model expects the retina images to be organized in a specific folder structure: `Dataset/train/class_name/image.png`. The images you uploaded are currently directly in the `/content/` directory. This step will create the necessary folders and move the images into their respective categories based on their filenames.

```python
import os
import shutil
from pathlib import Path

# Define the base directory for the dataset
DATASET_BASE_DIR = Path('/content/Dataset/train')

# Define the class names based on your file naming convention
CLASS_MAPPING = {
    'mil': 'mild',
    'mod': 'moderate',
    'severe': 'severe',
    'no_dr': 'no_dr',
    'no dr': 'no_dr', # Handle both 'no_dr' and 'no dr'
    'proliferate_dr': 'proliferate_dr'
}

# Create the necessary directory structure
for class_folder_name in set(CLASS_MAPPING.values()):
    (DATASET_BASE_DIR / class_folder_name).mkdir(parents=True, exist_ok=True)
print(f"Created dataset directory structure under: {DATASET_BASE_DIR}")

# Move images to their respective class folders
content_files = [f for f in Path('/content/').iterdir() if f.is_file() and f.suffix.lower() in ('.png', '.jpg')]

print("\nMoving images...")
for file_path in content_files:
    moved = False
    for keyword, class_name in CLASS_MAPPING.items():
        if keyword in file_path.name.lower():
            destination_folder = DATASET_BASE_DIR / class_name
            shutil.move(str(file_path), str(destination_folder / file_path.name))
            print(f"Moved '{file_path.name}' to '{destination_folder}'")
            moved = True
            break
    if not moved:
        print(f"Warning: '{file_path.name}' did not match any known class and was not moved.")

print("\nImage organization complete!")
```

### Step 2: Train the model and launch the Gradio demo

Now that the dataset is organized, we can run your provided script. This script will:
1.  Install all required Python packages.
2.  Load the dataset from the newly organized folders.
3.  Preprocess the images.
4.  Train the EfficientNet model through several phases (including fine-tuning).
5.  Evaluate the trained model and display performance metrics (accuracy, F1-score, confusion matrix).
6.  Save the trained model.
7.  Launch a Gradio web interface for you to upload new retina images and get predictions from the trained model. A shareable link will be provided by Gradio once it's launched.

```python
import subprocess
import sys
import os
from pathlib import Path
from PIL import Image
import numpy as np
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
import torch
from torch.optim.lr_scheduler import ReduceLROnPlateau
import timm
from collections import Counter
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, classification_report
import seaborn as sns
import gradio as gr
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

packages = [
    'torch',
    'torchvision',
    'timm',
    'opencv-python',
    'Pillow',
    'pandas',
    'scikit-learn',
    'matplotlib',
    'seaborn',
    'gradio'
]

print("Installing packages...")
for package in packages:
    try:
        subprocess.check_call([sys.executable, '-m', 'pip', 'install', package])
    except subprocess.CalledProcessError as e:
        print(f"Error installing {package}: {e}")
        sys.exit(1)

print("All packages installed successfully!")

# Define base dataset path
# IMPORTANT: Changed to absolute path for Colab environment
base_path = Path('/content/Dataset/train')

# Auto-discover classes from folder structure
print("Auto-discovering dataset classes...")
all_folders = [d for d in base_path.iterdir() if d.is_dir()]
classes = sorted([d.name for d in all_folders])

if not classes:
    print(f"✗ Error: No class folders found in {base_path}")
    sys.exit(1)

print(f"✓ Found {len(classes)} classes: {classes}")

# Check folder structure
print("\nChecking folder structure...")
for cls in classes:
    class_path = base_path / cls
    img_count = len(list(class_path.glob('*.jpg'))) + len(list(class_path.glob('*.JPG'))) + len(list(class_path.glob('*.png'))) + len(list(class_path.glob('*.PNG')))
    print(f"{cls}/: {img_count} images found")

# Load all images
print("\nLoading retina image data...")
data = {}
for cls in classes:
    class_path = base_path / cls
    images = []
    if class_path.exists():
        # Match both lowercase and uppercase extensions
        for img_file in sorted(list(class_path.glob('*.jpg')) + list(class_path.glob('*.JPG')) +
                                list(class_path.glob('*.png')) + list(class_path.glob('*.PNG'))):
            try:
                img = np.array(Image.open(img_file).convert('RGB'))
                images.append(img)
            except Exception as e:
                print(f"Error loading {img_file}: {e}")
    data[cls] = images
    print(f"Loaded {len(images)} images from {cls}")

print(f"\nDataset preparation complete! Total classes: {len(data)}")

# Display dataset summary
print("\nDataset Summary:")
for cls in classes:
    print(f"{cls}: {len(data[cls])} images")

# Verify image dimensions
print("\nImage dimensions:")
for cls in classes:
    if data[cls]:
        print(f"{cls}: {data[cls][0].shape}")

# Create a mapping of class names to indices for model training
class_to_idx = {cls: idx for idx, cls in enumerate(classes)}
idx_to_class = {idx: cls for cls, idx in class_to_idx.items()}

print(f"\nClass mapping: {class_to_idx}")
print("Dataset ready for model training!")

# Preprocess dataset
class RetinaDataset(Dataset):
    def __init__(self, data, class_to_idx, transform=None):
        self.data = data
        self.class_to_idx = class_to_idx
        self.transform = transform
        self.samples = []

        for cls, images in data.items():
            for img in images:
                self.samples.append((img, class_to_idx[cls]))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img, label = self.samples[idx]
        img = Image.fromarray(img.astype('uint8'))
        if self.transform:
            img = self.transform(img)
        return img, label

# Define preprocessing transforms
train_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomRotation(20),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# Create dataset and dataloader
dataset = RetinaDataset(data, class_to_idx, transform=train_transforms)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

print(f"Dataset preprocessed: {len(dataset)} samples")
print("Ready for model training with DataLoader!")

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Load pretrained EfficientNet model
model = timm.create_model('efficientnet_b0', pretrained=True, num_classes=len(classes))
model = model.to(device)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)

# Training function
def train_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0.0
    for images, labels in dataloader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(dataloader)

# Train model
num_epochs = 10
print("\nStarting model training...")
for epoch in range(num_epochs):
    avg_loss = train_epoch(model, dataloader, criterion, optimizer, device)
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}")
    scheduler.step(avg_loss)

print("Model training complete!")

# Fine-tuning strategy: Freeze backbone, train classifier
print("\nPhase 1: Training classifier with frozen backbone...")
for param in model.parameters():
    param.requires_grad = False

# The classifier layer name might vary by model. For EfficientNet_b0 it's `classifier`
# If it errors, inspect model.named_modules() to find the correct classifier name
for param in model.classifier.parameters():
    param.requires_grad = True

# Reinitialize optimizer for classifier-only training
optimizer = optim.Adam(model.classifier.parameters(), lr=0.001)

fine_tune_epochs = 5
for epoch in range(fine_tune_epochs):
    avg_loss = train_epoch(model, dataloader, criterion, optimizer, device)
    print(f"Fine-tune Epoch [{epoch+1}/{fine_tune_epochs}], Loss: {avg_loss:.4f}")
    scheduler.step(avg_loss)

# Phase 2: Unfreeze all layers for full fine-tuning
print("\nPhase 2: Full fine-tuning with all layers unfrozen...")
for param in model.parameters():
    param.requires_grad = True

optimizer = optim.Adam(model.parameters(), lr=0.0001)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)

final_epochs = 5
for epoch in range(final_epochs):
    avg_loss = train_epoch(model, dataloader, criterion, optimizer, device)
    print(f"Full Fine-tune Epoch [{epoch+1}/{final_epochs}], Loss: {avg_loss:.4f}")
    scheduler.step(avg_loss)

print("Fine-tuning complete!")
torch.save(model.state_dict(), 'retina_efficientnet_finetuned.pth')
print("Model saved as 'retina_efficientnet_finetuned.pth'")

# Enhanced preprocessing for retina images
def crop_retina_fundus(image):
    """Remove black/dark borders and crop to fundus region"""
    img_array = np.array(image)

    # Convert to grayscale for border detection
    if len(img_array.shape) == 3:
        gray = np.mean(img_array, axis=2).astype(np.uint8) # Ensure uint8 for consistent processing
    else:
        gray = img_array.astype(np.uint8)

    # Find non-black pixels (threshold at 30 to account for dark areas)
    threshold = 30
    non_black = gray > threshold

    # Find bounding box of fundus region
    rows = np.any(non_black, axis=1)
    cols = np.any(non_black, axis=0)

    if rows.any() and cols.any():
        ymin, ymax = np.where(rows)[0][[0, -1]]
        xmin, xmax = np.where(cols)[0][[0, -1]]

        # Crop with small margin
        margin = 5
        ymin = max(0, ymin - margin)
        xmin = max(0, xmin - margin)
        ymax = min(img_array.shape[0], ymax + margin)
        xmax = min(img_array.shape[1], xmax + margin)

        return Image.fromarray(img_array[ymin:ymax, xmin:xmax])

    return image

# Optimized transforms for EfficientNet (380x380)
enhanced_train_transforms = transforms.Compose([
    transforms.Lambda(crop_retina_fundus),
    transforms.Resize((380, 380)),
    transforms.RandomRotation(15),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.5),
    transforms.ColorJitter(brightness=0.15, contrast=0.15, saturation=0.1),
    transforms.GaussianBlur(kernel_size=3, sigma=(0.1, 1.0)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

val_transforms = transforms.Compose([
    transforms.Lambda(crop_retina_fundus),
    transforms.Resize((380, 380)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# Recreate dataset with enhanced preprocessing
dataset_enhanced = RetinaDataset(data, class_to_idx, transform=enhanced_train_transforms)
dataloader_enhanced = DataLoader(dataset_enhanced, batch_size=32, shuffle=True, num_workers=2)

print(f"Enhanced dataset preprocessed: {len(dataset_enhanced)} samples")
print("Retina fundus cropping and 380×380 resizing applied!")

# Check for class imbalance and compute weights
class_counts = Counter()
for cls, images in data.items():
    class_counts[cls] = len(images)

print("\nClass distribution:")
for cls in classes:
    print(f"{cls}: {class_counts[cls]} samples")

# Calculate class weights for imbalanced dataset
total_samples = sum(class_counts.values())
class_weights = torch.tensor([total_samples / (len(classes) * class_counts[cls])
                              for cls in classes], dtype=torch.float32).to(device)
print(f"Class weights: {class_weights}")

# Reinitialize model for fresh training
model = timm.create_model('efficientnet_b0', pretrained=True, num_classes=len(classes))
model = model.to(device)

# High performance settings
criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-5)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='min', factor=0.5, patience=5, min_lr=1e-7
)

# Early stopping parameters
best_loss = float('inf')
patience_counter = 0
max_patience = 10

# Training with validation function
def train_epoch_v2(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0.0
    correct = 0
    total = 0

    for images, labels in dataloader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        total_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

    avg_loss = total_loss / len(dataloader)
    accuracy = 100 * correct / total
    return avg_loss, accuracy

# PHASE 5: High performance training (20-50 epochs)
num_epochs = 30
print("\n" + "="*60)
print("PHASE 5: HIGH PERFORMANCE TRAINING")
print("="*60)

for epoch in range(num_epochs):
    avg_loss, accuracy = train_epoch_v2(model, dataloader_enhanced, criterion, optimizer, device)
    scheduler.step(avg_loss)

    print(f"Epoch [{epoch+1}/{num_epochs}] | Loss: {avg_loss:.4f} | Accuracy: {accuracy:.2f}%")

    # Early stopping
    if avg_loss < best_loss:
        best_loss = avg_loss
        patience_counter = 0
        torch.save(model.state_dict(), 'retina_best_model.pth')
        print(f"  ✓ Model improved! Saved checkpoint.")
    else:
        patience_counter += 1
        if patience_counter >= max_patience:
            print(f"\n✓ Early stopping triggered after {epoch+1} epochs")
            model.load_state_dict(torch.load('retina_best_model.pth'))
            break

print("\n" + "="*60)
print("HIGH PERFORMANCE TRAINING COMPLETE!")
print("="*60)

# Save final model
torch.save(model.state_dict(), 'retina_efficientnet_final.pth')
print("✓ Final model saved as 'retina_efficientnet_final.pth'")

# PHASE 6 — Evaluation
def evaluate_model(model, dataloader, device, classes):
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    return np.array(all_preds), np.array(all_labels)

# Evaluate on training data
print("\n" + "="*60)
print("PHASE 6: MODEL EVALUATION")
print("="*60)

predictions, labels = evaluate_model(model, dataloader_enhanced, device, classes)

# Calculate metrics
accuracy = accuracy_score(labels, predictions)
f1_macro = f1_score(labels, predictions, average='macro')
f1_weighted = f1_score(labels, predictions, average='weighted')
conf_matrix = confusion_matrix(labels, predictions)

print(f"\nOverall Accuracy: {accuracy:.4f}")
print(f"F1-Score (Macro): {f1_macro:.4f}")
print(f"F1-Score (Weighted): {f1_weighted:.4f}")

print("\nClassification Report:")
print(classification_report(labels, predictions, target_names=classes))

# Plot confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=classes, yticklabels=classes, cbar=True)
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')
print("\n✓ Confusion matrix saved as 'confusion_matrix.png'")
plt.close()

print("\n" + "="*60)
print("EVALUATION COMPLETE!")
print("="*60)

# PHASE 7 — Inference (User Upload Image)
def load_model_safe(model_path):
    """Load model with proper error handling"""
    # Check multiple possible locations
    possible_paths = [
        model_path,
        os.path.join(os.getcwd(), model_path),
        # Removed base_path.parent.parent as it's not relevant for loading from content/
    ]

    for path in possible_paths:
        if os.path.exists(path):
            print(f"✓ Model found at: {path}")
            model.load_state_dict(torch.load(path, map_location=device))
            return model

    raise FileNotFoundError(f"Model file not found in any of these locations:\n" + "\n".join(possible_paths))

def preprocess(image):
    """Preprocess a single image for inference"""
    if isinstance(image, str):
        image = Image.open(image).convert('RGB')
    elif isinstance(image, np.ndarray):
        image = Image.fromarray(image.astype('uint8'))

    return val_transforms(image).unsqueeze(0).to(device)

def predict(image):
    """Predict diabetic retinopathy stage from uploaded image"""
    model.eval()
    with torch.no_grad():
        img_tensor = preprocess(image)
        output = model(img_tensor)
        probabilities = torch.softmax(output, dim=1)
        pred = torch.argmax(probabilities, dim=1)
        confidence = probabilities[0][pred].item()

    predicted_class = idx_to_class[pred.item()]
    return f"{predicted_class} (Confidence: {confidence:.2%})"

# Load model before launching interface
try:
    load_model_safe('retina_efficientnet_final.pth')
    print("✓ Model loaded successfully for inference")
except FileNotFoundError as e:
    print(f"✗ Error: {e}")
    sys.exit(1)

# PHASE 8 — Local Testing UI (Gradio)
print("\n" + "="*60)
print("PHASE 8: LAUNCHING GRADIO INFERENCE UI")
print("="*60)

# Create Gradio interface
interface = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="pil", label="Upload Retina Image"),
    outputs=gr.Textbox(label="Prediction Result"),
    title="Diabetic Retinopathy Detection",
    description="Upload a retina fundus image to detect diabetic retinopathy stage",
    theme="default"
)
# Launch the interface
interface.launch(share=True)
```
